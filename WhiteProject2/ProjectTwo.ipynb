## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
import pandas as pd
fossil_ants = pd.read_csv("data/FossilTNRS.csv")
fossil_ants
import plotnine as p9

(p9.ggplot(data=fossil_ants,
           mapping=p9.aes(x='subfamily'))
    + p9.geom_bar()
    + p9.theme_bw()
    + p9.theme(axis_text_x = p9.element_text(angle=90))
    + p9.ggtitle("Subfamilies of Fossil Ants")
)
```

## Question 4: Perform a BLAST search of this ~unknown~ sequence (5 pts)

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

record = SeqIO.read("data/mysterSequence.fasta", format="fasta")
mystery_result = NCBIWWW.qblast("blastn", "nt", record.format("fasta"))
```

```python
save_file = open("results.xml", "w")
save_file.write(mystery_result.read())
save_file.close()
from Bio.Blast import NCBIXML

parsed = NCBIXML.read(open("results.xml"))
```
